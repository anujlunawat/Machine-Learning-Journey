```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
m = -2
b = 5
x = np.linspace(-3, 6, 10)
y = m*x + b
# plotting 4 points:
# 1. misclassified and near
# 2. misclassified and far
# 3. correctly classified and near
# 4. correctly classified and far

# the points are defined(below) in the same order as above i.e the first point is misclassified and near, and so on
# coordinates of the points:
x_points = (1.5, 6, 1.1, 5.4)
y_points = (3.5, 4, 4, 5.7)

plt.plot(x, y)
# Plot points
plt.plot(x_points[:2], y_points[:2], "ro", label="misclassified points")
plt.plot(x_points[2:], y_points[2:], "go", label="correctly classified points")

# Perpendicular slope
m_perp = -1 / m  # = 1/2

# Plot perpendicular lines
for xp, yp in zip(x_points, y_points):
    # Line through point with slope m_perp: y = m_perp*x + b
    b_perp = yp - m_perp * xp  # y - mx = b

    # Find intersection with original line: m*x + c = m_perp*x + b_perp
    x_intersect = (b_perp - c) / (m - m_perp)
    y_intersect = m * x_intersect + c

    # Draw line from point to intersection
    plt.plot([xp, x_intersect], [yp, y_intersect], 'k--', linewidth=1)



plt.ylim(0, 7.5)
plt.legend(loc="upper left")
plt.grid(True)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# for this, we introduce the sigmoid function
# sigmoid(z) = 1 / (1 + e^-z)

import math

x = np.linspace(-10, 10, 100)
y = 1 / (1 + math.e ** -x)

plt.plot(x, y)
plt.title('Sigmoid function')
plt.grid(True)
plt.show()
```

```python
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=30)
plt.scatter(X[:,0], X[:,1], c=y,s=100, cmap="winter")
```

```python
from sklearn.preprocessing import add_dummy_feature

class LogisticRegressionSigmoid:
    def __init__(self, epochs=100, learning_rate=0.1):
        self.coef_ = None
        self.epochs = epochs
        self.lr = learning_rate
    def fit(self, X, y):
        X = add_dummy_feature(X)
        self.coef_ = np.ones(X.shape[1])

        for _ in range(self.epochs):
            j = np.random.randint(X.shape[0])
            y_hat = LogisticRegressionSigmoid.sigmoid(np.dot(X[j], self.coef_))
            self.coef_ = self.coef_ + self.lr * (y[j] - y_hat) * X[j]
            
    @staticmethod
    def sigmoid(z):
        return 1 / (1 + np.exp(-z))
```

```python
lor = LogisticRegressionSigmoid(epochs=1000, learning_rate=0.1)
lor.fit(X, y)
```

```python
lor.coef_
```

```python
m = -(lor.coef_[1] / lor.coef_[2])
b = -(lor.coef_[0] / lor.coef_[2])

x_inp = np.linspace(-3, 3, 100)
y_inp = m*x_inp + b

plt.figure(figsize=(10,6))
plt.plot(x_inp, y_inp, "r-", linewidth=3)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="winter", s=100)
plt.ylim(-3,2)
plt.show()
```

Let's compare it with the sklearn LogisticRegression model

```python
from sklearn.linear_model import LogisticRegression
lor2 = LogisticRegression()
lor2.fit(X, y)
```

```python
m2 = -(lor2.coef_[0][0] / lor2.coef_[0][1])
b2 = -(lor2.intercept_ / lor2.coef_[0][1])

x_inp2 = np.linspace(-3, 3, 100)
y_inp2 = m2*x_inp2 + b2

plt.figure(figsize=(10,6))
plt.plot(x_inp, y_inp, "r-", linewidth=3, label="our model")
plt.plot(x_inp2, y_inp2, "b-", linewidth=3, label="sklearn model")
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="winter", s=100)
plt.ylim(-3,2)
plt.legend()
plt.show()
```
